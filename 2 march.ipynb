```python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

class Cow(Animal):
    def sound(self):
        print("Cow moos")

# Testing
animals = [Dog(), Cat(), Cow()]
for a in animals:
    a.sound()
```

```python
class Vehicle:
    def move(self):
        print("Vehicle moves")

class Car(Vehicle):
    def move(self):
        print("Car drives")

class Bike(Vehicle):
    def move(self):
        print("Bike rides")

class Bus(Vehicle):
    def move(self):
        print("Bus runs")

vehicles = [Car(), Bike(), Bus()]
for v in vehicles:
    v.move()
```

```python
class Payment:
    def pay(self, amount):
        pass

class CreditCardPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using Credit Card")

class JazzCashPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using JazzCash")

class BankTransferPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} via Bank Transfer")

payments = [CreditCardPayment(), JazzCashPayment(), BankTransferPayment()]
for p in payments:
    p.pay(1000)
```

```python
class Employee:
    def calculate_salary(self):
        pass

class FullTimeEmployee(Employee):
    def __init__(self, salary):
        self.salary = salary

    def calculate_salary(self):
        return self.salary

class PartTimeEmployee(Employee):
    def __init__(self, rate, hours):
        self.rate = rate
        self.hours = hours

    def calculate_salary(self):
        return self.rate * self.hours

class Intern(Employee):
    def __init__(self, stipend):
        self.stipend = stipend

    def calculate_salary(self):
        return self.stipend

emps = [
    FullTimeEmployee(50000),
    PartTimeEmployee(500, 40),
    Intern(10000)
]

for e in emps:
    print(e.calculate_salary())
```

```python
class Customer:
    def get_discount(self, amount):
        return amount

class RegularCustomer(Customer):
    def get_discount(self, amount):
        return amount * 0.97

class PremiumCustomer(Customer):
    def get_discount(self, amount):
        return amount * 0.92

class VIPCustomer(Customer):
    def get_discount(self, amount):
        return amount * 0.80

customers = [RegularCustomer(), PremiumCustomer(), VIPCustomer()]
for c in customers:
    print(c.get_discount(1000))
```

```python
class Account:
    def withdraw(self, amount):
        pass

class SavingsAccount(Account):
    def withdraw(self, amount):
        fee = amount * 0.02
        print("Withdraw:", amount - fee)

class CurrentAccount(Account):
    def withdraw(self, amount):
        print("Withdraw:", amount)

class BusinessAccount(Account):
    def withdraw(self, amount):
        fee = amount * 0.05
        print("Withdraw:", amount - fee)

accounts = [SavingsAccount(), CurrentAccount(), BusinessAccount()]
for acc in accounts:
    acc.withdraw(100)
```
