```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # Encapsulation (private variable)

    def deposit(self, amount):
        self.__balance += amount

    def show_balance(self):
        print("Balance is:", self.__balance)


class SavingsAccount(BankAccount):  # Inheritance
    def show_balance(self):          # Polymorphism
        print("Savings Balance is:", self._BankAccount__balance)


acc = SavingsAccount(1000)  # Object
acc.deposit(500)
acc.show_balance()
```

```python
class Car:
    def drive(self):
        print("Car is driving")


class SportsCar(Car):  # Inheritance
    def drive(self):    # Polymorphism
        print("Sports car is driving very fast ")


class Truck(Car):
    def drive(self):
        print("Truck is carrying heavy load ")


cars = [SportsCar(), Truck()]  # Objects

for car in cars:
    car.drive()
```

```python
class Student:
    def __init__(self, name, grade):
        self.name = name        
        self.grade = grade

    def introduce(self):
        print(f"My name is {self.name} and I study in grade {self.grade}")


class Monitor(Student):   # Inheritance
    def introduce(self):   # Polymorphism (overriding)
        print(f"I am {self.name}, class monitor of grade {self.grade}")


s1 = Student("Ali", 9)   # Object
s2 = Monitor("Ahmed", 9)

s1.introduce()
s2.introduce()
```
