```python
#oops 
class Dog:
    def sound(self):
        print("Dog barks")

class Cat:
    def sound(self):
        print("Cat meows")

class Cow:
    def sound(self):
        print("Cow moos")

animals = [Dog(), Cat(), Cow()]

for animal in animals:
    animal.sound()
```

```python
class IPCamera:
    def capture(self):
        print("Capturing via internet stream")

class CCTVCamera:
    def capture(self):
        print("Capturing via cable feed")

class DroneCamera:
    def capture(self):
        print("Capturing from aerial view")

cameras = [IPCamera(), CCTVCamera(), DroneCamera()]

for cam in cameras:
    cam.capture()
```

```python
#overiding
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")
```

```python
class Camera:
    def capture(self):
        print("Camera capturing...")


class IPCamera(Camera):
    def capture(self):
        print("IP Camera capturing via internet stream ")


class CCTVCamera(Camera):
    def capture(self):
        print("CCTV Camera capturing via cable ")


class DroneCamera(Camera):
    def capture(self):
        print("Drone Camera capturing from sky ")


cameras = [IPCamera(), CCTVCamera(), DroneCamera()]

for cam in cameras:
    cam.capture()
```
