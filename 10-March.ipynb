```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
df = pd.read_csv("d:\python\stock_dataset_dirty.csv")

df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df= df.sort_values("Date")
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()

df = df.drop_duplicates()
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Day_Name'] = df['Date'].dt.day_name()

df.head()
```

```python
# Average Closing Price by Company
import pandas as pd
df=pd.read_csv("d:\python\stock_dataset_dirty.csv")
group_symbol =df.groupby("Symbol")["Close"].mean()

print(group_symbol)
```

```python
sector_profit = df.groupby("Sector")["Profit_Margin"].mean()

print(sector_profit)
```

```python
volume_symbol = df.groupby("Symbol")["Volume"].sum()

print(volume_symbol)
```

```python
#Average Closing Price by Symbol and Sector
pivot_table = pd.pivot_table(df,
                             values="Close",
                             index="Symbol",
                             columns="Sector",
                             aggfunc="mean")

print(pivot_table)
```

```python
pivot_volume = pd.pivot_table(df,
                              values="Volume",
                              index="Symbol",
                              columns="Sector",
                              aggfunc="sum")

print(pivot_volume)
```
