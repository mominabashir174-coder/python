```python
# 1. Base class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 2. Method to display information
    def display_info(self):
        print("--- Person ---")
        print("Name:", self.name)
        print("Age:", self.age)


# 3. Derived class Student
class Student(Person):
    def __init__(self, name, age, student_id, major):
        super().__init__(name, age)
        self.student_id = student_id
        self.major = major

    # 5. Override display_info
    def display_info(self):
        print("--- Student ---")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Student ID:", self.student_id)
        print("Major:", self.major)


# 6. Derived class GraduateStudent
class GraduateStudent(Student):
    def __init__(self, name, age, student_id, major, thesis_topic):
        super().__init__(name, age, student_id, major)
        self.thesis_topic = thesis_topic

    # 7. Override display_info
    def display_info(self):
        print("\n--- Graduate Student ---")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Student ID:", self.student_id)
        print("Major:", self.major)
        print("Thesis Topic:", self.thesis_topic)


# 8. Create objects
person1 = Person("Momina", 20)
student1 = Student("saba", 21, "M567", "Computer Science")
grad_student1 = GraduateStudent("Maham", 27, "M314", "Physics", "Quantum Computing")

# Call display_info()
person1.display_info()
student1.display_info()
grad_student1.display_info()
```
