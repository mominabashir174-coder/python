```python
# 1. Use a for loop from 1 to 20
for num in range(1, 21):

    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)

# 3. Ask the user for a number
user_num = int(input("Enter a number for countdown: "))

# 5. Stop program if number is greater than 10
if user_num > 10:
    print("Number is greater than 10. Countdown stopped.")
else:
    # 4. Countdown using while loop
    while user_num >= 1:
        print("Countdown:", user_num)
        user_num -= 1

    print("Blast Off!")
```
