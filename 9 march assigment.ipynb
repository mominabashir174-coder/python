```python
# one value to multiple variables
# And you can assign the same value to multiple variables in one line
# 01
ali_fee = ahmed_fee = sara_fee = 5000
print(ali_fee)
print(ahmed_fee)
print(sara_fee)
# 02
mobile1_price = mobile2_price = mobile3_price = 30000
print("Mobile 1 Price:", mobile1_price)
print("Mobile 2 Price:", mobile2_price)
print("Mobile 3 Price:", mobile3_price)

# Unpacking a Collection
# If you have a collection of values in a list, tuple etc. Python allows you to extract the values into variables. This is called unpacking.
# 01
fruits = ["apple", "banana", "cherry"]
x, y, z = fruits
print(x)
print(y)
print(z)
# 02
car = ["Toyota", "Corolla", 2022]
brand, model, year = car
print(brand)
print(model)
print(year)

# Global Variables
# Global variable wo variable hota hai jo function ke bahar define kiya jata hai aur poore program me kahin bhi use ho sakta hai.
# 01
x = 10   # global variable
def my_function():
    print(x)
my_function()
# change the value of global variable
x = 5
def change_value():
    global x
    x = 20
change_value()
print(x)
# 02
school_name = "City School"   # global variable
def student_info():
    print("Student studies in", school_name)
def teacher_info():
    print("Teacher works at", school_name)
student_info()
teacher_info()
# Agar function me value change karni ho, to global keyword lagana padta hai.
# Global variable
balance = 1000
def deposit(amount):
    global balance   # function me global variable ko modify karne ke liye
    balance += amount
    print("Deposit Successful. New Balance:", balance)

def withdraw(amount):
    global balance
    balance -= amount
    print("Withdrawal Successful. New Balance:", balance)
# Function calls
deposit(500)
withdraw(200)
print("Final Balance:", balance)
```
