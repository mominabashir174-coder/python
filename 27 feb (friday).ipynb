```python
class Student:
    def __init__(self, name, grade):
        self.name = name        
        self.grade = grade

    def introduce(self):
        print(f"My name is {self.name} and I study in grade {self.grade}")


class Monitor(Student):   # Inheritance
    def introduce(self):   # Polymorphism (overriding)
        print(f"I am {self.name}, class monitor of grade {self.grade}")


s1 = Student("Ali", 9)   # Object
s2 = Monitor("Ahmed", 9)

s1.introduce()
s2.introduce()
```

```python
# Base Class
class Animal:
    def sound(self):
        print("Animal makes a sound")


# Child Class 1
class Dog(Animal):
    def sound(self):
        print("Dog barks ")


# Child Class 2
class Cat(Animal):
    def sound(self):
        print("Cat meows ")


# Child Class 3
class Cow(Animal):
    def sound(self):
        print("Cow moos ")


# Polymorphism in Action
animals = [Dog(), Cat(), Cow()]

for animal in animals:
    animal.sound()
```

```python
class Car:
    def drive(self):
        print("Car is driving")


class SportsCar(Car):  # Inheritance
    def drive(self):    # Polymorphism
        print("Sports car is driving very fast ")


class Truck(Car):
    def drive(self):
        print("Truck is carrying heavy load ")


cars = [SportsCar(), Truck()]  # Objects

for car in cars:
    car.drive()
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # Encapsulation (private variable)

    def deposit(self, amount):
        self.__balance += amount

    def show_balance(self):
        print("Balance is:", self.__balance)


class SavingsAccount(BankAccount):  # Inheritance
    def show_balance(self):          # Polymorphism
        print("Savings Balance is:", self._BankAccount__balance)


acc = SavingsAccount(1000)  # Object
acc.deposit(500)
acc.show_balance()
```

```python
class Bike:
    
    def __init__(self, brand, model, price):
        self.__brand = brand      # Private
        self.__model = model      # Private
        self._price = price       # Protected
    
    def get_brand(self):
        return self.__brand
    
    def set_price(self, new_price):
        if new_price > 0:
            self._price = new_price
        else:
            print("Price must be positive!")
    
    def display_info(self):
        print(f"Brand: {self.__brand}")
        print(f"Model: {self.__model}")
        print(f"Price: ${self._price}")


class SportsBike(Bike):
    
    def __init__(self, brand, model, price, top_speed):
        super().__init__(brand, model, price)
        self.__top_speed = top_speed
    
    def display_info(self):
        super().display_info()
        print(f"Top Speed: {self.__top_speed} km/h")


b1 = SportsBike("Yamaha", "R15", 2000, 150)
b1.display_info()
```
