```python
class Animal:
    def __init__(self, gender, breed):
        self.gender = gender
        self.breed = breed

# OUTSIDE the class
Animal1 = Animal("female", "husky")

print(Animal1.gender)
print(Animal1.breed)
```

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

car1 = Car("toyota", "corolla")

print(car1.brand)
print(car1.model)
```
